```python
import pandas as pd
import matplotlib.pyplot as plt  ## to plot
from matplotlib.pyplot import figure  ## to increase figure size 
 
 
 
df= pd.read_csv('Solar_profile.csv')  ## reading the file
df.columns = ['Time','Power']  ## changing column names to 'Time' and 'Power'
df['Time'] = df['Time'].apply(lambda x: x[11:25])  ## Excluding the dates and using only time stamps in 'Time' column
df = df.groupby('Time').sum()  ## Grouping same time stamps together irrespective of the dates
 
 
figure(figsize=(10, 8), dpi=80) ## Managing figure size 
plt.xticks(rotation=90)  ##Changing the x-label orintation to make it readable
plt.plot(df) ## Plotting the dataframe
plt.show()
```

**Conclusions:**
1. The Power curve of Solar Plant is kind of parabolic
2. Generation starts from approx 0530 hrs to 1830
3. Generation keeps on increasing from 0530 hrs till it reaches a peak at around 1130-1200 hrs and that is the time when solar radiations is at its peak
4. After the peak point, it again starts decreasing as the sun radiation starts reducing as it approaches towards the end of generation in the evening at around 1830 hrs.
